Preprocessing

```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# reading the dataset
data = pd.read_csv('online_shoppers_intention.csv')
# checking the shape of the data
print("Shape Of Data:",data.shape, end="\n\n")
# describing the data
print("Describe Data:")
print(data.describe(), end="\n\n")
# checking if the data contains any NULL values
print("No Of Null Values In Data:",data.isnull().sum().sum())
# Inputing Missing Values with 0
print("Filling Missing Values with 0")
data.fillna(0, inplace = True)
# checking the no. of null values in data after imputing the missing values
print("No Of Null Values In Data:",data.isnull().sum().sum(),"\n\n")
# one hot encoding
data1 = pd.get_dummies(data)
print("Columns in Data:",data1.columns)
# label encoding of revenue
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Revenue'] = le.fit_transform(data['Revenue'])
print("Revenue:\n",data['Revenue'].value_counts())
# getting dependent and independent variables
x = data1
# removing the target column revenue from x
x = x.drop(['Revenue'], axis = 1)
y = data['Revenue']
# checking the shapes
print("\n\nShape of x:", x.shape)
print("Shape of y:", y.shape)
```

Analysing Revenue on Weekend- Bivariate

```
df = pd.crosstab(data['Weekend'], data['Revenue'])
df.div(df.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True, figsize = (15, 5),
color = ['orange', 'crimson'])
plt.title('Weekend vs Revenue', fontsize = 30)
plt.show()
```

Analysing Revenue with respect to VisitorType- Bivariate


```
df = pd.crosstab(data['VisitorType'], data['Revenue'])
df.div(df.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True, figsize = (15, 5),
color = ['lightgreen', 'green'])
plt.title('Visitor Type vs Revenue', fontsize = 30)
plt.show()
```
